```python
def is_bbox_inside_image(bbox):
    x, y, w, h = bbox
    width, height = 320,320

    # Check if the bounding box is inside the image
    if x >= 0 and y >= 0 and (x + w) <= width and (y + h) <= height:
        return True
    else:
        return False

def is_bounding_box_outside_image(bounding_box):
  x, y, width, height = bounding_box
  if x < 0 or y < 0 or x + width > 320 or y + height > 320:
    return True
  else:
    return False

def move_bbox_inside_image(bbox, image_size):
    x, y, w, h = bbox
    width, height = image_size

    # Move the bounding box until it is inside the image
    while not is_bbox_inside_image(bbox):
        if x < 0:
            x = 0
        if y < 0:
            y = 0
        if (x + w) > width:
            x = width - w
        if (y + h) > height:
            y = height - h

        bbox = [x, y, w, h]

    return bbox
```

```python
new_anns = []
for ann in tqdm.tqdm(anns_0):
    bbox = ann['bbox']
    x = bbox[0]
    y = bbox[1]
    w = 10
    h = 10 


    ann['bbox'] = [int(x-w/2),int(y-h/2),w,h]
    ann['area'] = h*w


    if not is_bbox_inside_image(ann['bbox']):
        ann['bbox'] = move_bbox_inside_image(ann['bbox'],[320,320])
        ann['area'] = ann['bbox'][2]*ann['bbox'][3]

    new_anns.append(ann)



data['annotations'] = new_anns
print(data['annotations'])
```

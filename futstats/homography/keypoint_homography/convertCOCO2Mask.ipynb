```python
import cv2
import json
import matplotlib.pyplot as plt
import tqdm
from pycocotools.coco import COCO
```

```python
coco = COCO('/home/fer/Escritorio/futstatistics/datasets/dataset6_keypoints/dataset6_keypoints_coco/annotations/train_corrected_annotations.coco.json')
```

```python
img_coco = coco.imgs[0]

image = cv2.cvtColor(
    cv2.imread(f'/home/fer/Escritorio/futstatistics/datasets/dataset6_keypoints/dataset6_keypoints_coco/images/train/{img_coco["file_name"]}'),
    cv2.COLOR_BGR2RGB
)


plt.imshow(image)
```

```python
plt.imshow(image)
cat_ids = coco.getCatIds()
anns_ids = coco.getAnnIds(imgIds=img_coco['id'], catIds=cat_ids, iscrowd=None)
anns = coco.loadAnns(anns_ids)
print(anns)
coco.showAnns(anns)
```

```python
mask = coco.annToMask(anns[0])
for i in range(len(anns)):
    mask += coco.annToMask(anns[i])

plt.imshow(mask)
```
